```python
import cv2 
import numpy as np
import pandas as pd
from sklearn.utils import shuffle
img1 = cv2.imread('images/images/1.jpg') 
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img1 = np.array(img1)
img1 = img1.reshape([img1.shape[0] * img1.shape[1]])

img2 = cv2.imread('images/images/2.jpg') 
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
img2 = np.array(img2)
img2 = img2.reshape([img2.shape[0] * img2.shape[1]])

img3 = cv2.imread('images/images/3.jpg') 
img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2GRAY)
img3 = np.array(img3)
img3 = img3.reshape([img3.shape[0] * img3.shape[1]])

img4 = cv2.imread('images/images/4.jpg') 
img4 = cv2.cvtColor(img4, cv2.COLOR_BGR2GRAY)
img4 = np.array(img4)
img4 = img4.reshape([img4.shape[0] * img4.shape[1]])

riv = pd.read_csv('images/images/river.csv')
riv = shuffle(riv)
riv = riv[:50][:]
print(riv.head())
ri = []
for index, row in riv.iterrows():
    ri.append([ img1[(int(row['x'])* 512 )+ int(row['y'])] , img2[(int(row['x'])* 512 )+ int(row['y'])], img3[(int(row['x'])* 512 )+ int(row['y'])], img4[(int(row['x'])* 512 )+ int(row['y'])]])
ri = np.array(ri)
print(ri)
    
nriv = pd.read_csv('images/images/non_river.csv')
nriv = shuffle(nriv)
nriv = nriv[:100][:]
print(nriv.head())
nri = []
for index, row in nriv.iterrows():
    nri.append([ img1[(int(row['x'])* 512 )+ int(row['y'])] , img2[(int(row['x'])* 512 )+ int(row['y'])], img3[(int(row['x'])* 512 )+ int(row['y'])], img4[(int(row['x'])* 512 )+ int(row['y'])]])
nri = np.array(nri)

# print(ri)
mr = ri.sum(axis=0)/ri.shape[0]
mnr = nri.sum(axis=0)/nri.shape[0]
print(nri[10:20])
print(mnr)
print((nri - mnr)[10:20])
covar_r = ((ri - mr).T).dot(ri - mr)
covar_nr = ((nri - mnr).T).dot(nri - mnr)
print(covar_r)
print(covar_nr)
```

```python
test = np.array([img1,img2,img3,img4])
print(test.shape)
rtest = np.zeros(test.shape)
for i in range(test.shape[1]):
    rtest[0][i] = test[0][i] - mr[0]
    rtest[1][i] = test[1][i] - mr[1]
    rtest[2][i] = test[2][i] - mr[2]
    rtest[3][i] = test[3][i] - mr[3]

nrtest = np.zeros(test.shape)
for i in range(test.shape[1]):
    nrtest[0][i] = test[0][i] - mnr[0]
    nrtest[1][i] = test[1][i] - mnr[1]
    nrtest[2][i] = test[2][i] - mnr[2]
    nrtest[3][i] = test[3][i] - mnr[3]
#     print(nrtest[0][i])
print(rtest)
print(nrtest)
```
